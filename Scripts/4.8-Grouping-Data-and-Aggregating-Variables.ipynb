Part 2

```python
df_test = df_opm[:1000000]
```

```python
# aggregat of the first 1000000
df_test.groupby('department_id').agg({'number_of_consecutive_orders': ['mean']})
```

```python
# aggregat of the entire dataset
df_opm.groupby('department_id').agg({'number_of_consecutive_orders': ['mean']})
```

Part 3

For the most part both aggregated results are very similar. The bigest diffrence being that departments 3, 10, 12, and 21 all had there means shift by 1 or 2 whole numbers.

Part 4

```python
# Making the loyalty flag column
df_opm['max_order'] = df_opm.groupby(['user_id'])['number_of_consecutive_orders'].transform(np.max)
```

```python
df_opm.loc[df_opm['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_opm.loc[(df_opm['max_order'] <= 40) & (df_opm['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_opm.loc[df_opm['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_opm['loyalty_flag'].value_counts(dropna = False)
```

Part 6

```python
# Making the Spending flag column
df_opm['mean_spending'] = df_opm.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_opm.loc[df_opm['mean_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_opm.loc[df_opm['mean_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_opm['spending_flag'].value_counts(dropna = False)
```

Part 7

```python
# Making the order frequency flag column
df_opm['order_frequency'] = df_opm.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_opm.loc[df_opm['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_opm.loc[(df_opm['order_frequency'] <= 20) & (df_opm['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_opm.loc[df_opm['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_opm['order_frequency_flag'].value_counts(dropna = False)
```
