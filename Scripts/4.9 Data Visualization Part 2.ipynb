```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
path = r'C:\Users\aaron\OneDrive\Documents\CF Analysis 4'
```

```python
df_full = pd.read_pickle(os.path.join(path, '2 Data', 'Prepared Data', 'full_set'))
```

Part 2

```python
# Bar Chart
bar1 = df_full['orders_day_of_week'].value_counts().sort_index().plot.bar()
plt.title('Number of orders by day of week')
```

```python
# Export
bar1.figure.savefig(os.path.join(path, '4 Analysis','Visualizations', 'Bar_orders_day_of_week.png'), bbox_inches='tight')
```

```python
# Scatterplot
scatter1 = sns.scatterplot(x = 'prices', y = 'prices',data = df_full)
plt.title('Price differences (Scatterplot)')
```

```python
# Export
scatter1.figure.savefig(os.path.join(path, '4 Analysis','Visualizations', 'Scatter_prices_over_prices.png'), bbox_inches='tight')
```

```python
# Set up to make a Histogram
df_full.loc[df_full['prices'] >90, 'prices'] = np.nan
```

```python
# Histogram
his1 = df_full['prices'].plot.hist(bins = 25)
plt.title('Frequency of Prices')
```

```python
# Export
his1.figure.savefig(os.path.join(path, '4 Analysis','Visualizations', 'Histogram_prices.png'), bbox_inches='tight')
```

```python
his2 = df_full['prices'].plot.hist(bins = 70)
plt.title('Frequency of Prices')
```

```python
# Export
his2.figure.savefig(os.path.join(path, '4 Analysis','Visualizations', 'Histogram_prices_2.png'), bbox_inches='tight')
```
